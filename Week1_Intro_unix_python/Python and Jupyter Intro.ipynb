## What is Python?


<img src="https://www.notion.so/image/https%3A%2F%2Fm.media-amazon.com%2Fimages%2FM%2FMV5BMWY2ZGI0OGUtZDc3YS00ZmVjLWJiNWQtZDdmNzFmM2UzYWFhXkEyXkFqcGdeQXVyNTA4NzY1MzY%40._V1_.jpg?table=block&id=b138544c-366a-41e7-96cf-df7db0881794&spaceId=715086e5-0cab-4c71-8ba3-ab10dab963b1&width=2650&userId=&cache=v2" alt="Fun Fact" style="width: 30%; float:right;"/>

Python is a very-high level programming language that can be used for several computing tasks. These programs are written with a particular **syntax**. For example, to write or print my name on the screen, we can use the command `print('Chetan')`, written with the keyword print, the brackets (parenthesis) and the inverted commas as syntax. Python's syntax is verbose and simple, making it possible to write complex codes easily.
**Statement grouping** in Python is done using indentation and commands are separated by new lines. 

It is also an Object Oriented Programming Language (OOPS), so we can define certain Python objects in the code which takes on a set of properties making it reproducible and structured.

Python is an **interpreted** language, which is to say that a set of Python commands written together in a file to be executed as a program, or a Python script, is taken one line at a time, converted into code identifiable by our computer (machine code), and executed before going on to the next line in a top to bottom fashion. 

ðŸ’¡What then is the **difference between Unix scripts and Python scripts**? Unix Shell scripts are best at moving around files and changing text data due to their higher speeds as they interact directly with the Operating System. However, Python adds in several layers of abstractions, making it possible to write very complex scripts possible and easy. It is a very powerful language. It is also dynamically-typed, meaning that it designates the data types and structures automatically, without the need for us to tell them explicitly. All this makes Python program execution a bit slower than the same program in Unix, but far more functionally complex and diverse, yet programmatically easy.

# The Power of Python

```python
import pandas as pd #Import Pandas
import matplotlib.pyplot as plt #Import Matplotlib
stellar=pd.read_csv('closest20kstars.csv') #Read 20,000 Rows*97 columns of Data- Closest 20k stars to us as identified by Gaia
fig = plt.figure(figsize = [10,6]) #Make the figure
plt.scatter(stellar['ra'], stellar['dec'], alpha=0.8, s=0.7)
plt.xlabel('RA [$\degree$]')
plt.ylabel('Dec [$\degree$]')
plt.title('Closest Stars to us')
```

```python
stellar_trim=stellar[(stellar['radius_val'].notnull())]
#Magic Command for interactive 3D plots
%matplotlib notebook
fig = plt.figure(figsize = (10, 6)) #Make a figure
ax = plt.axes(projection ="3d") #Define 3D axes
scatter_plot=ax.scatter3D(stellar_trim['ra'], stellar_trim['dec'], stellar_trim['s_dist'], #Plot the scatter points
              s=stellar_trim['radius_val']*5, c=stellar_trim['abs_gmag'], cmap='RdYlBu', alpha=0.7) #Style scatter points
ax.set_xlabel('RA [$\degree$]') #Set labels
ax.set_ylabel('Dec [$\degree$]')
ax.set_zlabel('Distance [pc]')
plt.title('Closest Stars with known radius') #Set title
fig.colorbar(scatter_plot, label="absolute G-band photometric magnitude") #Get colorbar in the figure
```
